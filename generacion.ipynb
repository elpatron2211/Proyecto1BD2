```python
import random
import pandas as pd
```

```python
#Empresas

"""
Props:

Tipo: Fabricante, Proveedor
Nombre: Nombre de la empresa
Pais: Pais de la empresa
Telefono: Telefono de la empresa
Email: Email de la empresa

"""



def generar_empresas(cantidad):
    empresas = [["Tipo", "Nombre", "Pais", "Telefono", "Email"]]
    tipos = ['Fabricante', 'Proveedor']
    paises = ['Estados Unidos', 'Mexico', 'Canada', 'Japon', 'China', 'Alemania', 'Francia', 'Italia', 'Brasil', 'India']
    emails = ['.com', '.net']
    for i in range(cantidad):

        empresas.append(
            [random.choice(tipos),
            f'Empresa {i}',
            random.choice(paises),
            f'{random.randint(1000000000, 9999999999)}',
            f'compras@empresa{i}{random.choice(emails)}']
            )
        
    return empresas

#Productos
"""
Props:

Nombre: Nombre del producto
Precio: Precio del producto
Codigo: Codigo del producto
Tipo: Tipo de producto
Marca: Marca del producto

"""

def generar_producto(cantidad):
    productos = [["Nombre", "Precio", "Codigo", "Tipo", "Marca"]]
    
    marcas = ['Apple', 'Samsung', 'LG', 'Sony', 'Microsoft', 'HP', 'Dell', 'Asus', 'Acer', 'Lenovo']
    tipos = ['Computadora', 'Tablet', 'Smartphone', 'Smartwatch', 'Consola', 'Audifonos', 'Television', 'Impresora', 'Proyector', 'Camara']

    for i in range(cantidad):
        productos.append([
            f'Producto {i}',
            f'{random.randint(100, 9999)}',
            f'{random.randint(1000000000, 9999999999)}',
            random.choice(tipos),
            random.choice(marcas)]
        )

    return productos

#Documentos:

"""
Props:
TipoDocumento: Tipo de documento
FechaEmisión: Fecha del documento
Numero: Numero del documento
Estado: Estado del documento
Total: Total del documento

"""



def generar_documentos(cantidad):
    tipos_documentos = ['Orden de Compra', 'Despacho a Tienda', 'Transferencia a Almacen']
    estados = ['Pendiente', 'Completado']
    fechas = pd.date_range(start='1/1/2020', end='1/1/2024', freq='D')
    documentos = [["TipoDocumento", "FechaEmisión", "Numero", "Estado", "Total"]]

    for i in range(cantidad):
        documentos.append([
            random.choice(tipos_documentos),
            random.choice(fechas),
            i,
            random.choice(estados),
            f'{random.randint(100, 9999)}']
        )

    return documentos

#Almacenes

"""
Props:

Nombre: Nombre del almacen
Direccion: Direccion del almacen
Telefono: Telefono del almacen
Ciudad: Ciudad del almacen
TipoAlmacen: Tipo de almacen

"""



def generar_almacenes(cantidad):
    tipos_almacenes = ['Almacen Recepción', 'Almacen Despacho']
    ciudades = ['CDMX', 'Guadalajara', 'Monterrey', 'Tijuana', 'Puebla', 'Merida', 'Cancun', 'Acapulco', 'Veracruz', 'Toluca']
    almacenes = [["Nombre", "Direccion", "Telefono", "Ciudad", "TipoAlmacen"]]
    for i in range(cantidad):
        almacenes.append(
            [f'Almacen {i}',
            f'Calle {random.randint(1, 100)}',
            f'{random.randint(1000000000, 9999999999)}',
            random.choice(ciudades),
            random.choice(tipos_almacenes)]
        )

    return almacenes

#Tiendas

"""
Props:

Nombre: Nombre de la tienda
Direccion: Direccion de la tienda
Telefono: Telefono de la tienda
Ciudad: Ciudad de la tienda
Asesor: Gerente de la tienda

"""


def generar_tiendas(cantidad):
    gerentes = ['Juan', 'Pedro', 'Maria', 'Jose', 'Ana', 'Luis', 'Carlos', 'Fernando', 'Ricardo', 'Roberto']
    ciudades = ['CDMX', 'Guadalajara', 'Monterrey', 'Tijuana', 'Puebla', 'Merida', 'Cancun', 'Acapulco', 'Veracruz', 'Toluca']
    tiendas = [["Nombre", "Direccion", "Telefono", "Ciudad", "Asesor"]]
    for i in range(cantidad):
        tiendas.append(
            [f'Tienda {i}',
            f'Calle {random.randint(1, 100)}',
            f'{random.randint(1000000000, 9999999999)}',
            random.choice(ciudades),
            random.choice(gerentes)]
        )

    return tiendas

#Camiones

"""
Props:

Placas: Placas del camion
Marca: Marca del camion
Modelo: Modelo del camion
Capacidad: Capacidad del camion
Piloto: Piloto del camion

""" 


def generar_camiones(cantidad):
    marcas_camiones = ['Kenworth', 'Volvo', 'Freightliner', 'International', 'Peterbilt', 'Mack', 'Western Star', 'Hino', 'Isuzu', 'Ford']
    camiones = [["Placas", "Marca", "Modelo", "Capacidad", "Tipo"]]
    pilotos = ['Juan', 'Pedro', 'Maria', 'Jose', 'Ana', 'Luis', 'Carlos', 'Fernando', 'Ricardo', 'Roberto']
    placas = random.sample(range(100000, 999999), cantidad)
    for i in range(cantidad):
        camiones.append(
            [placas[i],
            random.choice(marcas_camiones),
            f'{random.randint(2000, 2022)}',
            f'{random.randint(1000, 9999)}',
            random.choice(pilotos)]
        )

    return camiones







```

```python
def DISTRIBUYE_A(empresas):
    fabricantes = []
    proveedores = []
    for empresa in empresas:
        if empresa[0] == 'Fabricante':
            fabricantes.append(empresa)
        else:
            proveedores.append(empresa)

    print(fabricantes)
    print(proveedores)
    fechas_inicio = pd.date_range(start='1/1/2015', end='1/1/2020', freq='D')
    fechas_fin = pd.date_range(start='1/1/2021', end='1/1/2030', freq='D')
    distribuyeA = [['fabricante', 'proveedor', 'fechaInicioContrato', 'fechaFinContrato', 'precioContrato']]
    for i in fabricantes:
        numero_proveedores = random.randint(1, 5)
        for j in range(numero_proveedores):
            FechaInicioContrato = random.choice(fechas_inicio)
            FechaFinContrato = random.choice(fechas_fin)
            precioContrato = random.randint(10000, 100000)
            distribuyeA.append([i[1], random.choice(proveedores)[1], FechaInicioContrato, FechaFinContrato, precioContrato])



    return distribuyeA

def DISTRIBUIDO_POR(empresas, productos):
    distribuidos = [['empresa', 'producto', 'costo', 'horasTomaPedidos', 'distribuidoDesde']]
    horasTomaPedidos = ['7-13','8-17','8-15', '10-17','9-18']
    proveedores = []
    fechas = pd.date_range(start='1/1/2015', end='1/1/2020', freq='D')
    for empresa in empresas:
        if empresa[0] == 'Proveedor':
            proveedores.append(empresa)
    for i in productos:
        distribuidos.append([random.choice(proveedores)[1], i[0], round(random.uniform(1, 50),2),random.choice(horasTomaPedidos), random.choice(fechas)])
    return distribuidos

def FABRICADO_POR(empresas, productos):
    fabricados = [['empresa', 'producto', 'materiales', 'costo', 'fechaInicio']]
    fabricantes = []
    materiales = ['Plastico', 'Aluminio', 'Fibra de Carbono', 'Silicón', 'Cobre', 'Acero', 'Cerámica', 'Cobre', 'Grafeno', 'Níquel', 'Cobalto']
    fechas = pd.date_range(start='1/1/2015', end='1/1/2020', freq='D')
    for empresa in empresas:
        if empresa[0] == 'Fabricante':
            fabricantes.append(empresa)
    for i in productos:
        prod_mats = []
        for j in range(random.randint(2,5)):
            material = random.choice(materiales)+' '+str(j+1)
            prod_mats.append(material)
        fabricados.append([random.choice(fabricantes)[1], i[0], prod_mats,round(random.uniform(50, 150),2), random.choice(fechas)])
    return fabricados

def INVENTARIO(almacenes, tiendas, productos):
    inventario_almacen = [['almacen', 'producto', 'cantidad', 'shelfTimePromedioDias','bloque']]
    inventario_tienda = [['tienda', 'producto', 'cantidad', 'shelfTimePromedioDias', 'pasillo']]
    for i in almacenes:
        for j in productos:
            inventario_almacen.append([i[0], j[0], random.randint(0, 100), round(random.uniform(1,30), 2), random.choice(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])+'-'+str(random.randint(1,8))])

    for i in tiendas:
        for j in productos:
            inventario_tienda.append([i[0], j[0], random.randint(0, 10), round(random.uniform(15,60),2), random.randint(1,12)])
    return inventario_almacen, inventario_tienda

def DESPACHOSyENTREGAS(camiones, documentos, almacenes, tiendas):
    almacenesDes = [i for i in almacenes if i[4] == 'Almacen Despacho']
    despacho_a =  [['almacen', 'camion', 'documento', 'codigoConfirmacion', 'fechaEntrega']]
    completado_por = [['documento', 'camion', 'comision','ayudante','tiempoDescargaMinutos']]
    entregado_en = [['camion', 'tienda', 'documento', 'integridadProductos', 'tiempoDescargaMinutos']]
    solicitud = [['documento', 'tienda', 'firmadaPor', 'modalidad', 'metodoPago']]
    ayudantes = ['Juan', 'Pedro', 'Maria', 'Jose', 'Ana', 'Luis', 'Carlos', 'Fernando', 'Ricardo', 'Roberto']
    modalidades = ['App', 'Correo Electrónico', 'Llamada Telefónica', 'Visita presencial']
    metodospagos = ['Efectivo', 'Tarjeta de Crédito', 'Tarjeta de Débito', 'Transferencia Bancaria', 'Cheque']
    for i in documentos:
        if i[0] == 'Despacho a Tienda' and i[3] == 'Completado':
            almacen = random.choice(almacenesDes)
            camion1 = random.choice(camiones)
            camion2 = random.choice(camiones)
            tienda = random.choice(tiendas)
            despacho_a.append([almacen[0], camion1[0], i[2], random.randint(100000,999999), i[1]+pd.Timedelta(hours=random.randint(12, 48))])
            completado_por.append([i[2], camion1[0], round(random.uniform(100,250),2), random.choice(ayudantes), random.randint(30, 120)])
            entregado_en.append([camion2[0], tienda[0], i[2], str(random.randint(90,100))+'%', random.randint(30, 120)])
            solicitud.append([i[2], tienda[0], random.choice(ayudantes), random.choice(modalidades), random.choice(metodospagos)])
    return despacho_a, completado_por, entregado_en, solicitud

def ORDENES_DE_COMPRA_y_TRASFERENCIA(documentos, empresas, almacenes):
    empresaProv = [x for x in empresas if x[0] == 'Proveedor']
    ordenes = [['documento', 'almacen', 'hechaPor', 'hora', 'responsable']]
    ordenado_a_empresa = [['documento', 'empresa', 'corresponsal', 'numeroConfirmacion', 'prioridad']]
    ordenado_a_almacen = [['documento', 'almacen', 'administrador','numeroConfirmacion', 'prioridad']]
    for i in documentos:
        if i[0] == 'Orden de Compra':
            ordenes.append([i[2], random.choice(almacenes)[0], 'Empleado '+str(random.randint(1,100)), random.randint(8,18), 'Gestor '+str(random.randint(1,5))])
            ordenado_a_empresa.append([i[2], random.choice(empresaProv)[1], 'Corresponsal '+str(random.randint(1,100)), random.randint(100000,999999), random.choice(['Alta', 'Media', 'Baja'])])
        elif i[0] == 'Transferencia a Almacen':
            almacen1 = random.choice(almacenes)
            almacen2 = random.choice(almacenes)
            while almacen1 == almacen2:
                almacen2 = random.choice(almacenes)
            ordenes.append([i[2], almacen1[0], 'Empleado '+str(random.randint(1,100)), random.randint(8,18), 'Gestor '+str(random.randint(1,5))])
            ordenado_a_almacen.append([i[2], almacen2[0], 'Administrador '+str(random.randint(1,25)), random.randint(100000,999999), random.choice(['Alta', 'Media', 'Baja'])])
    return ordenes, ordenado_a_empresa, ordenado_a_almacen

def PRODUCTOS_ORDEN(documentos, productos):
    productosOrden = [['documento', 'producto', 'cantidad', 'colores', 'Premium']]
    colores = ['Negro', 'Gris', 'Plateado', 'Dorado', 'Azul', 'Rojo', 'Multicolor']
    for i in documentos:
        numero_prods = random.randint(1, 10)
        for j in range(numero_prods):
            productosOrden.append([i[2], random.choice(productos)[0], random.randint(1, 10), random.choice(colores), random.choice(True, False)])
    return productosOrden

def DISTANCIAS(almacenes, tiendas):
    distancias = [['almacen', 'tienda', 'distancia', 'tiempo_mins', 'costo']]
    almacenesDes = [i for i in almacenes if i[4] == 'Almacen Despacho']
    for i in almacenesDes:
        for j in tiendas:
            distancias.append([i[0], j[0], random.randint(1, 100), random.randint(1, 100), random.randint(5, 500)])

    return distancias



def extract_csv(almacenes, camiones, documentos, empresas, productos, tiendas):
    distribuye_a = DISTRIBUYE_A(empresas)
    distribuido_por = DISTRIBUIDO_POR(empresas, productos)

    fabricado_por = FABRICADO_POR(empresas, productos)
    inventario_almacen, inventario_tienda = INVENTARIO(almacenes, tiendas, productos)
    despachos, completados, entregados, solicitudes = DESPACHOSyENTREGAS(camiones, documentos, almacenes, tiendas)
    ordenes, ordenado_a_empresa, ordenado_a_almacen = ORDENES_DE_COMPRA_y_TRASFERENCIA(documentos, empresas, almacenes)
    productosOrden = PRODUCTOS_ORDEN(documentos, productos)
    distancias = DISTANCIAS(almacenes, tiendas)

    pd.DataFrame(distribuye_a[1:],columns=distribuye_a[0]).to_csv('distribuye_a.csv', index=False)
    pd.DataFrame(distribuido_por[1:], columns=distribuido_por[0]).to_csv('distribuido_por.csv', index=False)
    pd.DataFrame(fabricado_por[1:], columns= fabricado_por[0]).to_csv('fabricado_por.csv', index=False)
    pd.DataFrame(inventario_almacen[1:], columns=inventario_almacen[0]).to_csv('inventario_almacen.csv', index=False)
    pd.DataFrame(inventario_tienda[1:], columns=inventario_tienda[0]).to_csv('inventario_tienda.csv', index=False)
    pd.DataFrame(despachos[1:], columns=despachos[0]).to_csv('despachos.csv', index=False)
    pd.DataFrame(completados[1:], columns=completados[0]).to_csv('completados.csv', index=False)
    pd.DataFrame(entregados[1:], columns=entregados[0]).to_csv('entregados.csv', index=False)
    pd.DataFrame(solicitudes[1:], columns = solicitudes[0]).to_csv('solicitudes.csv', index=False)
    pd.DataFrame(ordenes[1:], columns=ordenes[0]).to_csv('ordenes.csv', index=False)
    pd.DataFrame(ordenado_a_empresa[1:], columns=ordenado_a_empresa[0]).to_csv('ordenado_a_empresa.csv', index=False)
    pd.DataFrame(ordenado_a_almacen[1:], columns=ordenado_a_almacen[0]).to_csv('ordenado_a_almacen.csv', index=False)
    pd.DataFrame(productosOrden[1:], columns=productosOrden[0]).to_csv('productosOrden.csv', index=False)
    pd.DataFrame(distancias[1:], columns=distancias[0]).to_csv('distancias.csv', index=False)
```

```python
def main(num_almacenes, num_camiones, num_documentos, num_empresas, num_productos, num_tiendas):
    almacenes = generar_almacenes(num_almacenes)
    camiones = generar_camiones(num_camiones)
    documentos = generar_documentos(num_documentos)
    empresas = generar_empresas(num_empresas)
    productos = generar_producto(num_productos)
    tiendas = generar_tiendas(num_tiendas)

    pd.DataFrame(almacenes[1:], columns=almacenes[0]).to_csv('almacenes.csv', index=False)
    pd.DataFrame(camiones[1:], columns=camiones[0]).to_csv('camiones.csv', index=False)
    pd.DataFrame(documentos[1:], columns=documentos[0]).to_csv('documentos.csv', index=False)
    pd.DataFrame(empresas[1:], columns=empresas[0]).to_csv('empresas.csv', index=False)
    pd.DataFrame(productos[1:], productos[0]).to_csv('productos.csv', index=False)
    pd.DataFrame(tiendas[1:], tiendas[0]).to_csv('tiendas.csv', index=False)

    extract_csv(almacenes[1:], camiones[1:], documentos[1:], empresas[1:], productos[1:], tiendas[1:])

main(50, 100, 4000, 200, 2500, 500)
```
